```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from  sympy.solvers import ode
import sympy as sp
import math as m
import matplotlib.animation as anim
```

```python
t3=np.linspace(0,100,10000)
```

```python
figurap1=plt.figure(figsize=(6,5))
axp1= figurap1.add_subplot(111)
def System(r0,t):
    dydt=r0[0]+2*r0[1]
    dxdt=2*r0[0]-r0[1]
    return [dydt,dxdt]

for i in range(-10,10):
    for j in range(-10,10):
        r0=[i,j]
        soli=odeint(System,r0,t3)
        plt.plot(soli[:,0],soli[:,1])
axp1.set_xlim(-20,20)
axp1.set_ylim(-15,15)
axp1.scatter(0,0,color="k",label="Punto critico",marker=".")
axp1.legend()
```
